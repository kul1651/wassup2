* 주제 : 온라인 학습 플랫폼 사용자 행동 패턴 분석
* 개요 :
```
이번 프로젝트는 학습 서비스를 이용하는 독자들 중 누가 서비스 구독을 연장할지 예측하기 위한 사전 파악에 초점을 맞추고 있습니다. 우리의 목표는 유저들의 학습 습관, 선호도, 이용 행태 등 다양한 측면을 면밀히 분석하는 것입니다.


이를 위해, 우리는 유저들의 학습 패턴, 서비스 이용 시간, 콘텐츠에 대한 반응, 그리고 이들이 얼마나 자주 서비스를 이용하는지 등 다양한 데이터를 수집하고 분석할 계획입니다. 이 데이터는 유저들이 서비스에 얼마나 만족하고 있는지, 어떤 요소들이 그들의 구독 결정에 영향을 미치는지 이해하는 데 중요한 역할을 할 것입니다.
```
* 데이터셋 : mini.csv
    * 메타데이터
        * user_id: 사용자의 고유 식별자
        * subscription_duration: 사용자가 서비스에 가입한 기간 (월)
        * recent_login_time: 사용자가 마지막으로 로그인한 시간 (일)
        * average_login_time:  사용자의 일반적인 로그인 시간
        * average_time_per_learning_session: 각 학습 세션에 소요된 평균 시간 (분)
        * monthly_active_learning_days: 월간 활동적인 학습 일수
        * total_completed_courses: 완료한 총 코스 수
        * recent_learning_achievement: 최근 학습 성취도
        * abandoned_learning_sessions: 중단된 학습 세션 수
        * community_engagement_level: 커뮤니티 참여도
        * preferred_difficulty_level: 선호하는 난이도
        * subscription_type: 구독 유형
        * customer_inquiry_history: 고객 문의 이력
        * payment_pattern : 사용자의 지난 3개월 간의 결제 패턴을 10진수로 표현한 값.
          - 7: 3개월 모두 결제함
          - 6: 첫 2개월은 결제했으나 마지막 달에는 결제하지 않음
          - 5: 첫 달과 마지막 달에 결제함
          - 4: 첫 달에만 결제함
          - 3: 마지막 2개월에 결제함
          - 2: 가운데 달에만 결제함
          - 1: 마지막 달에만 결제함
          - 0: 3개월 동안 결제하지 않음
        * target: 사용자가 다음 달에도 구독을 계속할지 (1) 또는 취소할지 (0)를 나타냄

## 배경 확인 및 가설 세우기

현재상황 파악  
1. 메인타겟(주 고객층 선별) -> 2. 메인타겟의 구매패턴 및 구독 유형 -> 3. 주 타겟의 구독 연장 이유를 파악 (부족한 부분은 보완하고, 잘하고 있는 분야는 유지 및 발전) 

* 문제 파악: 서비스 구독 연장자 확보(고객확보)를 위함(불만족 사항을 개선하여 구독 연장을 늘리기 위함) -> "리텐션"

* 요구사항 분석 
구매패턴 및 최근 학습 성취도와의 상관관계  

---

* 가설: 1. 일단 주 고객층을 선별하고 -> 그들의 학습 성취도, 선호도, 구매 패턴을 통한 구독 연장 이유를 먼저 파악해야 할 것, 

 2. 학습태도가 좋고, 커뮤니티 참여도가 높은(적극성) 교육생일수록 구독 연장 확률이 높을 것,

 3. 구독 유형을 통한 사용자 선호도 -> 다음 구독 연장에도 영향을 미칠 것(학습목적), 

 4. 구독결졍에 미치는 영향: 새로운것을 배우고 싶거나, 부족한 공부를 보강하기 위해 서비스 이용, 기존 콘텐츠 만족도 또한 영향을 미칠것, 학업 성취도(최근 학습 성취도가 높은 구독자) 또한 수강 동기 및 유지를 위해 중요한 포인트

---
* 원하는 결과: 서비스 구독을 연장 할지 예측 -> 서비스 연장 가능성이 높은 교육생을 확보 함과 동시에, 구독 연장 가능성이 있는 구독자에게는 혜택 및 개선을 통한 구독 연장 유도 및 구독 연장 지표를 높이고 싶음 -> 구독 연장 교육생을 늘림과 동시에 유지하기 위함

- 커뮤니티 참여도가 활발한 사람은 구독연장에 영향을 줄것이라 예측했었는데

- 중단된 학습 세션은 2-3개인 수강생이 가장 많았으며, 중단된 학습 세션이 수강 연장에 영향을 줄것으로 예상했으나, 이는 큰 영향이 없는것으로 나타났습니다. 오히려 중단된 학습 세션을 완료하기 위해 수강연장을 하는것이 아닌지 살펴볼 필요가 있습니다.
